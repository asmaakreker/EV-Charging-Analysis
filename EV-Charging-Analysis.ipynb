```python
# for data handling we import pandas library
import pandas as pan 
#for plotting graphs we import matplotlib
import matplotlib.pyplot as vsp 
# here for database operations we import pymongo 
import pymongo
#here we import collections module for advanced data structure
import collections
#importing statistical functions
import scipy.stats
#for advanced visualizations we import seaborn
import seaborn as sbsb
#importing numpy for the numerical operations 
import numpy as nnpy
```

```python
#reading the data from the csv file 
ev_charging = pan.read_csv("ev_charging_station_usage.csv")
```

```python
#here we display general info about the dataset using info()
ev_charging.info()
```

```python
#for loop to loop through all categorical columns (object data type)
for data in ev_charging.select_dtypes(include=['object']):
    #calculate and print the percentage distribution of each category in the column
    print((ev_charging.groupby(data).size()/ev_charging[data].count())*100)
```

```python
#here we are dropping the 'Station_ID' column because it is not needed for analysis
ev_charging =ev_charging.drop(['Station_ID'], axis=1)
```

```python
# here we are checking if there are any missing values 
ev_charging.isnull().sum()
```

```python
# we define a function to convert the categorical columns into numerical values
def encode_cat_to_num(ev_charging, ccv): 
    #for loop to loop through each column in the list
    for ae in ccv: 
        #just checking if the column exists in the dataset
        if ae in ev_charging.columns:  
            #here we get the unique values of the column  
            the_unique_var = ev_charging[ae].unique()
            #here we replace every unique value with a number 
            ev_charging[ae].replace(the_unique_var, range(len(the_unique_var)), inplace=True)
```

```python
# write a list containing all the columns, dont include station_id becuase it is dropped
ccv= [ 'City', 'Date', 'Vehicle_Type', 'Charging_Start_Time', 'Charging_End_Time', 'Energy_Consumed_kWh', 'Cost_INR', 'Payment_Method']
#converting categorical data into numerical values
encode_cat_to_num(ev_charging,ccv)
```

```python
#displaying the first 5 rows after encodig 
ev_charging.head() 
```

```python
#here we calculate the mean of the Cost_INR column 
the_mean = ev_charging['Cost_INR'].mean()
#here we print the mean value
print("mean:" + str(the_mean))

```

```python

#finding the max of the Cost_INR
the_max = ev_charging['Cost_INR'].max()
# printing the max value
print("max:" +  str(the_max))
```

```python
#finding the min of Cost_INR
the_min = ev_charging['Cost_INR'].min()
#we print the min value
print("min:" +  str(the_min))
```

```python
#here we calculate the correlation of all the variables with Cost_INR
corr_numbers = ev_charging.corr()['Cost_INR'] 
#printing the correlation values
print(corr_numbers)
```

```python
#use the spearmanr() method to perform the Spearman correlation test between Vehicle_Type and Cost_INR
scipy.stats.spearmanr(ev_charging['Vehicle_Type'], ev_charging['Cost_INR']) 
```

```python
#here we create a pivot table with mean, min, and max of Cost_INR grouped by Vehicle_Type
tabular_summ_pivot = ev_charging.pivot_table(
    values='Cost_INR',
    index='Vehicle_Type', 
    aggfunc=[nnpy.mean, nnpy.min, nnpy.max] 
)
#printing the pivot table
print(tabular_summ_pivot)
```

```python
#here we create another pivot table using sum aggregation 
ev_charging.pivot_table(index=['Vehicle_Type'], aggfunc=nnpy.sum)
```

```python
#here we are creating another pivot table grouping by Vehicle_Type and Cost_INR, then sum all other columns 
ev_charging.pivot_table(index=['Vehicle_Type', 'Cost_INR'], aggfunc=nnpy.sum)
```

```python
#creating a cross-tabulation between Vehicle_Type and Cost_INR
pan.crosstab(ev_charging['Vehicle_Type'], ev_charging['Cost_INR'], margins=True)
```

```python
#here we are setting the figure size for the plots 
vsp.figure(figsize=(10,6))
#to get the correlation values by using the corr function
corr_val = ev_charging.corr()
# create a box plot using the seaborn library 
sbsb.heatmap(corr_val, annot=True)
#adding title to the plot 
vsp.title('Correlation of EV charging cars')
#lable the x-axis
vsp.xlabel('Type of vehicle')
#lable the y-axis
vsp.ylabel('Cost')
#use show() to display the plot 
vsp.show()
```

```python
#here we create a scatter plot between Vehicle_Type and Cost_INR
vsp.figure(figsize=(8,6))

vsp.scatter(
    ev_charging['Vehicle_Type'],
    ev_charging['Cost_INR']
)
#here we are adding title to the plot
vsp.title('cost vs. type')
#lable the x-axis
vsp.xlabel('type')
#lable the y-axis
vsp.ylabel('cost')
#use show() to display the plot
vsp.show()
```

```python
#in this line we create a pie chart of Cost_INR distribution
vsp.title('Cost_INR')
ev_charging['Cost_INR'].value_counts().plot.pie()
```

```python
#Here we create a boxplot to compare cost across vehicle types
sbsb.boxplot(x='Vehicle_Type', y='Cost_INR', data=ev_charging)
#add a title to the boxplot
vsp.title('How Diff Vehicles Differ In Cost')
#set the x-axis and y-axis lables
vsp.xlabel('Type ')
vsp.ylabel('cost ')
#using show() to display the boxplot
vsp.show()
```

```python
#In this line we create a bar chart showing average cost per vehicle type
ev_charging.groupby('Vehicle_Type')['Cost_INR'].mean().plot(kind='bar')
#set a title to the bar chart
vsp.title('How Diff Vehicles Differ In Cost')  
#setting the x and y axis lable 
vsp.xlabel('Type')
vsp.ylabel('Cost') 
# show the bar chart 
vsp.show()
```
